```python
# modules we'll use
import numpy as np
import pandas as pd
import datetime
```

```python
# read in all our data
df = pd.read_csv('DataFrame.csv')
```

```python
# look at the first five rows of the data file
df.head()
```

```python
# get the number of missing data points per column
missing_values_count = df.isnull().sum()
```

```python
# look at the number of missing points in the first ten columns
missing_values_count[0:10]
```

```python
# how many total missing values do we have?
total_cells = np.product(df.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

```python
# remove all columns with at least one missing value
columns_with_na_dropped = df.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
df = columns_with_na_dropped
```

```python
# convert our date columns to datetime
import datetime
df['Date'] = pd.to_datetime(df['Date'] , format = "%Y%m%d")
```

```python
df["Date"]=pd.to_datetime(df["Date"].astype("str"))
dept_year=[]
dept_month=[]
dept_date = []
for i in df['Date'].astype("str"):
    k=i.strip()
    dept_year.append(k.split("-")[0])
    dept_month.append(k.split("-")[1])
    dept_date.append(k.split("-")[2])
df["YEAR"] = dept_year
df["MONTH"] = dept_month
df["DATES"] = dept_date
```
